```python
from textblob import TextBlob
```

```python
def get_sentiment(text):
    blob=TextBlob(text)
    sentiment=blob.sentiment.polarity
    if sentiment>0:
     result = "positive"
    elif sentiment<0:
     result = "negative"
    else:
     result = "neutral"
    return result
```

```python
get_sentiment("i love you")
```

```python
get_sentiment("i hate you")
```

```python
get_sentiment("i will remember the matter ")
```

```python
df['sentiment']=df['Text'].apply(get_sentiment)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,plot_confusion_matrix
```

```python
from sklearn.model_selection import train_test_split
```

```python
import neattext.functions as nfx
dir(nfx)
df['clean_text']=df['Text'].apply(nfx.remove_stopwords)
df['clean_text']=df['Text'].apply(nfx.remove_userhandles)
df['clean_text']=df['Text'].apply(nfx.remove_punctuations)
```

```python
x1=df['clean_text']
y1=df['Emotion']
```

```python
x1
```

```python
cv=CountVectorizer()
x=cv.fit_transform(x1)
```

```python
cv.get_feature_names()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y1,test_size=0.3,random_state=42)
```

```python
nv=MultinomialNB()
nv.fit(x_train,y_train)
nv.score(x_test,y_test)
```

```python
y_pred_nv=nv.predict(x_test)
```
